```python
#1)	Import pyplot as  plt from matplotlib
from matplotlib import pyplot as plt	
```

```python
#2)	Create an inline graph by passing data which has 5 pairs and show the graph
x=[0,5,10,15,20]
y=[10,15,20,20,30]
plt.plot(x,y,'-.',color='y',linewidth=3.5,marker='D',markersize=7,markerfacecolor='blue',markeredgecolor='red',markeredgewidth=1.5)
plt.show()
```

```python
#3)	Add label to both axes and provide a title and show the graph
x=[0,5,10,15,20]
y=[10,15,20,20,30]
plt.plot(x,y,'-.',color='y',linewidth=3.5,marker='D',markersize=7,markerfacecolor='aqua',markeredgecolor='red',markeredgewidth=1.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('LINE GRAPH')
plt.show()
```

```python
#4)	Change the colour of lines in multiline inline graph
x=[0,5,10,15,20]
y=[10,15,20,20,30]
y1=[20,30,45,55,50]
plt.plot(x,y,'-.',color='maroon',linewidth=3.5,marker='D',markersize=7,markerfacecolor='blue',markeredgecolor='red',markeredgewidth=1.5,label='X- value')
plt.plot(x,y1,'--',color='hotpink',linewidth=3.5,marker='D',markersize=7,markerfacecolor='blue',markeredgecolor='red',markeredgewidth=1.5,label='Y- value')
plt.title('LINE GRAPH')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
	
```

```python
#5)	Create a bar graph with title for axes, graph, legends present, specify own margins and then show it
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],label="BMW",width=.5,color='hotpink')
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],label="Audi", color='purple',width=.5)

plt.legend()
plt.title("CARS PERFORMANCE")
plt.xlabel("Days")
plt.ylabel("Distance covered in kms")

plt.show()
```

```python
#6)	Create a histogram with title for axes, graph, legends present, specify own margins and then show it.
import numpy as np
age=np.random.randint(0,40,20)
print(age)
bins= [0,10,20,30,40]
plt.hist(age,bins,histtype='barstacked',rwidth=0.5,color='purple',density=True)
plt.title("HISTOGRAMS")
plt.xlabel("age groups")
plt.ylabel("no.of.people")
plt.show()	
```

```python
#7)	Create a scatterplot with title for axes, graph, legends present, specify own margins and then show it.
x = [1,1.5,2,2.5,3,3.5,3.6]
y = [7.5,8,8.5,9,9.5,10,10.5]
x1=[8,8.5,9,9.5,10,10.5,11]
y1=[3,3.5,3.7,4,4.5,5,5.2]
plt.scatter(x,y, label='high income low saving',color='r',marker='s',linewidth=2,edgecolor='black')
plt.scatter(x1,y1,label='low income high savings',color='yellow',marker='o',linewidth=2,edgecolor='purple')
plt.xlabel('saving*100')
plt.ylabel('income*1000')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
#8)	Create an area plot with title for axes,graph,legends present specify own margins and then show it

days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]
plt.plot([],[],color='black',label='sleeping',linewidth=5)
plt.plot([],[],color='b',label='eating',linewidth=5)
plt.plot([],[],color='hotpink',label='working',linewidth=5)
plt.plot([],[],color='g',label='playing',linewidth=5)
plt.stackplot(days,sleeping,eating,working,playing,colors=['black','b','hotpink','g'],baseline='weighted_wiggle')
plt.title('Activities')
plt.xlabel('Days')
plt.ylabel('Hours spent')
plt.legend()
```

```python
#9)	create a pie chart with relevant details

slices=[7,2,2,13]
activities=['sleeping','eating','working','playing']
col=['c','purple','linen','hotpink']
plt.pie(slices,labels=activities,colors=col,shadow=True,explode=(0,0.1,0,0),autopct='%1.3f%%',startangle=90)
plt.title('Daily Activities')
plt.show()
```
