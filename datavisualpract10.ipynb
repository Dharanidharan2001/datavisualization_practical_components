```python
""" 6. Label the axes and add a title
    7. Arrange borders and grid lines
    8. Use gridspec to partition the figure into subplots.
    9. Adjust the dots transparency
    10. Colour the dots using your own colour palette and python colour maps
    11. Iterate through dataset inorder to assign a label and colour to each data point. 
"""
import matplotlib.cm as cm
def scatterplot(df, x_dim, y_dim, category):
    x = df[x_dim]
    y = df[y_dim]
    categories = df[category].apply(lambda x: 'maximum wickets taken' if x > 20 else 'minimum wickets taken')
    fig, ax = plt.subplots(figsize=(10, 5))
    colors = ['#2300A8', '#00A658']
    for i in range(len(df)):
        ax.scatter(x.loc[i], y.loc[i], alpha=0.70, color = colors[i%len
        (colors)], label=categories.loc[i])
        ax.set_title('Total overs played Vs wicket takens')
        ax.set_xlabel('total overs Played')
        ax.set_ylabel('madins')
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
        ax.legend(categories.unique())

scatterplot(df, 'Overs','Mdns', 'Wkts')
```
